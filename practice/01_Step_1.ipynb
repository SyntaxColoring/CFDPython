```python
# Boilerplate

import numpy
from matplotlib import pyplot
import time, sys
%matplotlib inline
```

```python
# Input

max_x = 2
num_points = 41
point_interval = max_x / num_points

num_timestamps = 40
time_interval = 0.025

wave_speed = 1
```

```python
# Initial state

initial_velocities = numpy.ones(num_points)
initial_velocities[int(len(initial_velocities)*0.125):int(len(initial_velocities)*0.5)] = 2

print(initial_velocities)
pyplot.plot(numpy.linspace(0, max_x, num_points), initial_velocities)
```

```python
# Simulation

current_velocities = initial_velocities
for timestamp in range(num_timestamps):
    next_velocities = numpy.empty(num_points)
    next_velocities[0] = current_velocities[0]
    for i, current_velocity in enumerate(current_velocities):
        if i == 0:
            continue
        next_velocities[i] = current_velocities[i] - wave_speed*time_interval/point_interval * (current_velocities[i] - current_velocities[i-1])
    current_velocities = next_velocities

pyplot.plot(numpy.linspace(0, 2, num_points), current_velocities);
```
